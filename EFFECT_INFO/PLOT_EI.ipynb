## Notebook to plot EI and check convergence

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df1 = pd.read_table('3x3_sims_cold2.txt',skiprows=[0,1,2])
df4 = pd.read_table('3x3_sims_cold.txt',skiprows=[0,1,2])
df2 = pd.read_table('3x3_sims_crit.txt',skiprows=[0,1,2])
df3 = pd.read_table('3x3_sims_hot.txt',skiprows=[0,1,2])
#df5 = pd.read_table('4x4_sims_hot.txt',skiprows=[0,1,2])

plt.plot(df1['Steps'],df1['EI'],color='green',alpha=0.6,label='3x3 Cold')
plt.plot(df4['Steps'],df4['EI'],color='orange',alpha=0.6,label='3x3 Cold 2')
plt.plot(df2['Steps'],df2['EI'],color='blue',alpha=0.6,label='3x3 Critical')
plt.plot(df3['Steps'],df3['EI'],color='purple',alpha=0.6,label='3x3 Hot')
#plt.plot(df5['Steps'],df5['EI'],color='red',alpha=0.6,label='4x4 Hot')

#plt.xlim((0,2300000))
plt.xlabel('Steps')
plt.ylabel('Effective Info (EI)')
plt.title('Effective Information')
plt.legend(loc='best')
#plt.savefig('/Users/jakehanson/Desktop/3x3_EI.pdf')
plt.show()
```
