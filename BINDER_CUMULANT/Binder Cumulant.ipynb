## Notebook to calculate the binder cumulant as a function of T and lattice size in order to benchmark ising model
- Should cross near critical temperature

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Init arrays
T_array = [2.0,2.05,2.1,2.15,2.2,2.25,2.3,2.35]
n_array = [16,32,64,128,256]

for n in n_array:
    U_array = []
    print 'n = ',n
    for T in T_array:
        path = 'MAGNETIZATION/n'+str(n)+'.T'+str(T)+'.txt'
        data = pd.read_table(path,header=None,names=['Magnetization'])
        M = data['Magnetization']
        U = 1-np.mean(M**4)/(3*np.mean(M**2)**2)
        U_array.append(U)
    plt.plot(T_array,U_array,label='n = '+str(n),alpha=0.6)
plt.xlabel('Temperature')
plt.ylabel('Binder Cumulant U4')
plt.axvline(2.269185,label=r'Critical Temp',alpha=0.5,color='black')
plt.xlim(2.0,3.0)
plt.legend()
plt.savefig('/Users/jakehanson/Desktop/binder_cumulant.pdf')
plt.show()
```
